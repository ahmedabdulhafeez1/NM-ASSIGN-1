## Bivariate Analysis

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the file and its path
data= pd.read_csv('/content/House Price India.csv')
```

```python
#Shape
data.shape
```

```python
#Datatypes in the CSV file
data.dtypes
```

```python
#Descriptioin of the CSV
data.describe()
```

```python
#Checking the missing values in the dataset.
data.isnull().sum()
```

```python
#ScatterPlot
sns.scatterplot(x='Date', y='Price', data=data)
```

```python
#line plot
sns.lineplot(x='Date', y='Price', data=data)
```

```python
#box plot
sns.boxplot(x='Date', y='Price', data=data)
```

```python
#statistical analysis
from scipy.stats import pearsonr
correlation, p_value = pearsonr(data['Date'], data['Price'])
print('Correlation coefficient:', correlation)
print('p-value:', p_value)
```

```python
#Head
data.head()
```

## Univariate Analysis

```python
#Plotting the histogram
sns.histplot(data['Price'], kde=True)
```

```python
#Box Plot
sns.boxplot(data['Price'])
```

```python
#Violin Plot
sns.violinplot(data['Price'])
```

```python
#statistical analysis
from scipy.stats import ttest_1samp
t_statistic, p_value = ttest_1samp(data['Price'], 0)
print('t-statistic:', t_statistic)
print('p-value:', p_value)
```

## Multivariate Analysis

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
```

```python
sns.pairplot(data)
```
